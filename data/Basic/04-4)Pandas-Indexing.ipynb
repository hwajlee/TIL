### - 인덱싱 예제

```python
# 새로운 데이터프레임 생성
data = {'Name':['Chulmin','Eunkyoung','Jiwoong','Soobeom'],
        'Year':[2011,2016,2015,2015],
        'Gender':['Male','Female','Male','Male'],
        }

data_df = pd.DataFrame(data, index = ['one', 'two', 'three', 'four'])
data_df
```

```python
# 인덱싱 시작 1
data_df[['Name','Gender']]

# data_df[0] - 단일값으로 행과 열 모두 인덱싱 안됨
data_df[0:3]
#행은 슬라이싱으로만 된다. 
```

```python
# 단일값을 불러오기 1
data_df.loc['three','Name']
```

```python
# 단일값을 불러오기 2
data_df.iloc[2,0]
```

```python
data_df.loc['one':'two','Name':'Year']
```

```python
data_df.loc[data_df.Year>2014][:2]
```

###  - boolean indexing

```python
titan_df = pd.read_csv('C:/Users/kh983/Projects/TIL/2. Big Data/Problems/3) titanic/train.csv')
titan_df
```

```python
#1. 60세가 넘는 사람불러오기
titan_df[titan_df.Age>60].reset_index(drop=1).head(3)
# old인덱스와 함께 불러와진다.
```

```python
#2. 성별이 여성인 승객조건을 만족하는 '이름', '나이'컬럼 추출하기
titan_df[titan_df.Sex=='female'][['Name','Age',]]
```

```python
#3. 객실등급이 1등급인 사람들을 추출해보자
titan_df[titan_df.Pclass==1].Pclass
# titan_df[titan_df.Pclass==1].Pclass.value_counts()
```

```python
# 위 세 조건을 결합하기
titan_df[~(titan_df.Age > 60) | ~(titan_df.Sex == 'female') | ~(titan_df.Pclass == 1)]
```

```python
titan_df[(titan_df.Age <= 60) | (titan_df.Sex != 'female') | (titan_df.Pclass != 1)]
```
