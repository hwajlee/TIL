```python
# 방법1
chipo_df['new_price'] = chipo_df.item_price.str[1:].astype('float64')
```

```python
## q9. 데이터 프레임 new_price 값이 new_price값의 평균값 이상을 가지는 데이터들을 인덱싱해주세요.
```

```python
## q10. 데이터 프레임 item_name에서 '6 Pack Soft Drink'의 값을 'pepsi'로 변형해주세요.
```

```python
# 방법1 - lambda를 활용한 식
chipo_df.item_name.apply(lambda x : 'pesci' if x=='6 Pack Soft Drink' else x)
```

```python
# 방법2 - loc를 활용
chipo_df.loc[chipo_df.item_names=='6 Pack Soft Drink', 'item_names'] = 'pepsi'
```

```python
# 방법3 - replace 함수 활용
chipo_df['item_name'] = chipo_df['item_name'].str.replace('6 Pack Soft Drink','pepsi')
```

```python
chipo_df.item_name.value_counts()
```
