### 여러 그래프

```python
# y1 = 2x^2
# y2 = 5x + 30
# y3 = 4x^2 + 10 

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-4.5, 5.0, 0.5)

y1 = 2*x**2
y2 = 5*x + 30 
y3 = 4*x**2 + 10 
```

```python
### 여러개의 함수 그래프 그리기 - 1

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.show()
```

```python
### 여러개의 함수 그래프 그리기 - 2

plt.plot(x, y1, x, y2, x, y3)
plt.show()
```

```python
### 새로운 창에 그래프를 생성해보자 

plt.plot(x, y1) # 처음 그림이 나타남 

### 새 창을 추가 
plt.figure(figsize=(10,6)) # 새로운 창 추가 
plt.plot(x, y2) # 새로운 직선이 그려진다
plt.show()
```

```python
### 1번 데이터쌍(x,y1) 2번 데이터쌍(x,y2)를 지정된 번호로 
# 그래프 창에 입력해보자. 새로운 창에 그래프를 생성해보자

### y1 = x^2 - 2 
### y2 = 20 * cos(x)^2
import numpy as np

# 데이터 생성 
x = np.arange(-5, 5, 0.1)
y1 = x**2 - 2 
y2 = 20*np.cos(x)**2

plt.figure(0) # 0번 그래프 창에 넣어주세요 
plt.plot(x, y1)


plt.figure(1) # 1번 그래프 창에 넣어주세요 
plt.plot(x, y2)

plt.figure(2) # 2번 그래프 창에 넣어주세요 
plt.plot(x, y2)
```
